```python
from pygbif import occurrences as occ
from pygbif import registry
import pandas as pd
import os
import time
import matplotlib.pyplot as plt
```

```python
# Let's find all the files in the GBIF database related to Seaflower
resultsSeaflower =  registry.dataset_search('seaflower')['results']
resultsSanAndres =  registry.dataset_search('san andrés')['results']
resultsProvidencia =  registry.dataset_search('providencia')['results']
print(len(resultsSeaflower), len(resultsSanAndres), len(resultsProvidencia))
results = [resultsSeaflower, resultsSanAndres, resultsProvidencia]
```

```python
results
```

```python
# Now, let's extract the name and the key for each Seaflower dataset and save them into a dictionary

def generate_ECS_df(dict_search_gbif):
    """
    input: a dictionary provided by the search made in GBIF
    output: the dictionary search is converted into a dataframe
    """    
    proyectos_filtrados = []
    dictKeysSeaflower = {}
    for dic in dict_search_gbif:

        title = dic['title']
        try:
            description = dic['description']
        except:
            description = ''

        try:
            recordCount = dic['recordCount']
        except:
            recordCount = ''

        try:
            projectIdentifier = dic['projectIdentifier']
        except:
            projectIdentifier = ''

        try:
            publishingOrganizationTitle = dic['publishingOrganizationTitle']
        except:
            publishingOrganizationTitle = ''

        try:
            hostingOrganizationTitle = dic['hostingOrganizationTitle']
        except:
            hostingOrganizationTitle = ''
        key = dic['key']
        if 'seaflower' in title.lower() or 'sea flower' in title.lower() or 'serranilla' in title.lower() \
            or 'san andrés' in title.lower() or 'providencia' in title.lower():
            dictKeysSeaflower[dic['key']] = [title,
                                             description, 
                                             recordCount,
                                            projectIdentifier,
                                            publishingOrganizationTitle,
                                             hostingOrganizationTitle, key]
        else:
             proyectos_filtrados.append(title)
            
    df_projects = pd.DataFrame(list(dictKeysSeaflower.values()), columns= ['projectTitle',
                                                                       'description',
                                                                     'recordCount',
                                                                     'projectIdentifier',
                                                                     'publishingOrganizationTitle',
                                                                     'hostingOrganizationTitle','datasetKey'])
    df_projects['Downloaded'] = [0]*len(df_projects)
    print('Lista de proyectos filtrados:')
    for proyecto in proyectos_filtrados:
        print(proyecto)
    return df_projects
    
providencia_df = generate_ECS_df(resultsProvidencia)
sanAndres_df = generate_ECS_df(resultsSanAndres)
seaFlower_df = generate_ECS_df(resultsSeaflower)

```

```python
seaFlower_df
```

```python
df_ECS = pd.merge(seaFlower_df, providencia_df,how = 'outer', on=list(seaFlower_df.columns))
filter_ECS = ['Colombia_BIO','ECS']
bool_list = []
for id_project in df_ECS['projectIdentifier']:
    split = id_project.split(' ')
    contador_False = 0
    for element in split:
        if element in filter_ECS:
            bool_list.append(True)
            break
        else:
            contador_False += 1
        if contador_False == len(split):
            bool_list.append(False)

# df_ECS[df_ECS['projectIdentifier'].isin(filter_ECS)]
```

```python
len(bool_list),len(df_ECS)
```

```python
df_ECS_final = df_ECS[bool_list].reset_index(drop = True)
for description in df_ECS_final['description']:
    print(description,'\n')
```
